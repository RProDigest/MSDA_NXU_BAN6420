## Data Visualizations

### Plot of income by Age

```python
# Sort the DataFrame by income to highlight the highest incomes
df_sorted = df.sort_values(by="Income", ascending=False)

# Bar plot of Income by Age
sns.barplot(x="Age", y="Income", data=df_sorted)
plt.title("Ages with the Highest Income")
plt.savefig("income_by_age_distribution.png", format="png", dpi=300, bbox_inches="tight") #Export for use in Powerpoint
plt.show()
```

### Plot of gender distribution across spending categories

```python
# We melt the DataFrame to make 'Spending_Category' a variable
df_melted = df.melt(id_vars=["Gender"], 
                    value_vars=["Utilities", "Entertainment", "School_Fees", "Shopping", "Healthcare"],
                    var_name="Spending_Category", value_name="Amount_Spent")


# Bar plot to show the average spending by gender across categories
sns.barplot(x="Spending_Category", y="Amount_Spent", hue="Gender", data=df_melted, estimator=sum)
plt.title("Gender Distribution Across Spending Categories")
plt.xticks(rotation=45)  # Rotate the x labels for better readability
plt.savefig("gender_distribution_spending_categories.png", format="png", dpi=300, bbox_inches="tight") #Export for use in Powerpoint
plt.show()
plt.show()
```
